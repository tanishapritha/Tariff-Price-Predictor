```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set plot style
sns.set(style="whitegrid")

# Load datasets
tariffs = pd.read_csv("data/tariff_data.csv")
prices = pd.read_csv("data/Prices_E_All_Data_NOFLAG.csv")
inflation = pd.read_csv("data/inflation.csv")
deflation = pd.read_csv("data/deflation.csv")
gdp = pd.read_csv("data/gdppcap.csv")
indices = pd.read_csv("data/FAOSTAT_data_en_4-12-2025.csv")  # NEW CSV

# Preview datasets
print("Tariffs:")
display(tariffs.head())

print("\nPrices:")
display(prices.head())

print("\nInflation:")
display(inflation.head())

print("\nDeflation:")
display(deflation.head())

print("\nGDP per capita:")
display(gdp.head())

print("\nFood Price Indices (FAO):")
display(indices.head())
print("\nColumns in Food Price Indices CSV:")
print(indices.columns)

# Optional: Check for missing values in each dataset
print("\nMissing values in each dataset:")
for name, df in zip(["Tariffs", "Prices", "Inflation", "Deflation", "GDP", "Indices"],
                    [tariffs, prices, inflation, deflation, gdp, indices]):
    print(f"{name}: {df.isnull().sum().sum()} missing values")
```

```python
# Check shapes
print("Shapes of datasets:")
for name, df in zip(["Tariffs", "Prices", "Inflation", "Deflation", "GDP", "Indices"],
                    [tariffs, prices, inflation, deflation, gdp, indices]):
    print(f"{name}: {df.shape}")

# Optional: Inspect year ranges and country names
print("\nYears in each dataset (if applicable):")
print("Tariffs years:", [col for col in tariffs.columns if col.isdigit()])
print("GDP years:", [col for col in gdp.columns if col.isdigit()])
print("Inflation years:", [col for col in inflation.columns if col.isdigit()])
print("Deflation years:", [col for col in deflation.columns if col.isdigit()])
print("Indices years:", indices['Year'].unique() if 'Year' in indices.columns else "No 'Year' column")
```

```python
tariffs_long = tariffs.melt(id_vars=['Entity', 'Code'], var_name='Year', value_name='Tariff Rate')
tariffs_long['Year'] = tariffs_long['Year'].astype(int)  # Convert Year to integer
```

```python
# Check unique values in the Year column to see if there's anything odd
print(tariffs['Year'].unique())
print(gdp['Year'].unique())
print(inflation['Year'].unique())
print(deflation['Year'].unique())
print(indices['Year'].unique())
```

```python
# Transpose the inflation, deflation, and GDP datasets so that years become rows
inflation_transposed = inflation.T
inflation_transposed.columns = ['Inflation']
inflation_transposed.reset_index(inplace=True)
inflation_transposed.rename(columns={'index': 'Year'}, inplace=True)

deflation_transposed = deflation.T
deflation_transposed.columns = ['Deflation']
deflation_transposed.reset_index(inplace=True)
deflation_transposed.rename(columns={'index': 'Year'}, inplace=True)

gdp_transposed = gdp.T
gdp_transposed.columns = ['GDP']
gdp_transposed.reset_index(inplace=True)
gdp_transposed.rename(columns={'index': 'Year'}, inplace=True)

# Clean up price-e-noflag column names and extract the year
prices.columns = [col[1:] if col.startswith('Y') else col for col in prices.columns]  # Remove 'Y' from column names

# Now we have all datasets in a form where 'Year' is a column
```
