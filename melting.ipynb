```python
import pandas as pd

# Load the dataset
gdp = pd.read_csv("data/gdp-pcap.csv")

# Identify year columns (all numeric column names)
year_cols = [col for col in gdp.columns if col.isdigit()]

# All other columns are ID columns
id_cols = [col for col in gdp.columns if col not in year_cols]

# Melt the dataframe to long format
gdp_long = gdp.melt(id_vars=id_cols, var_name="Year", value_name="GDP_per_capita")

# Convert Year to integer
gdp_long["Year"] = gdp_long["Year"].astype(int)

# Optional: View result
print(gdp_long.head())
```

```python
import pandas as pd

# Load the inflation dataset
inflation = pd.read_csv("data/inflation.csv")

# Identify year columns (all numeric column names)
year_cols = [col for col in inflation.columns if col.isdigit()]

# All other columns are ID columns
id_cols = [col for col in inflation.columns if col not in year_cols]

# Melt the dataframe
inflation_long = inflation.melt(id_vars=id_cols, var_name="Year", value_name="Inflation")

# Convert 'Year' to integer
inflation_long["Year"] = inflation_long["Year"].astype(int)

# Preview
print(inflation_long.head())
```

```python
import pandas as pd

# Load the deflation dataset
deflation = pd.read_csv("data/deflation.csv")

# Identify year columns (those that are digits)
year_cols = [col for col in deflation.columns if col.isdigit()]

# Identify id columns (the rest)
id_cols = [col for col in deflation.columns if col not in year_cols]

# Melt the dataframe into long format
deflation_long = deflation.melt(id_vars=id_cols, var_name="Year", value_name="Deflation")

# Convert 'Year' to integer
deflation_long["Year"] = deflation_long["Year"].astype(int)

# Preview
print(deflation_long.head())
```

```python
import pandas as pd

# Load the FAO dataset
fao = pd.read_csv("data/fao-data.csv")

# Select only useful columns
fao_clean = fao[[
    "Area", "Year", "Item", "Element", "Unit", "Value"
]].copy()

# Rename for consistency with other datasets
fao_clean = fao_clean.rename(columns={
    "Area": "Country Name"
})

# Optional: Clean up countries (if needed) or drop missing
fao_clean = fao_clean.dropna(subset=["Country Name", "Year", "Value"])

# Preview
print(fao_clean.head())
```

```python
import pandas as pd

# Load the FAO dataset
fao_data = pd.read_csv("data/fao-data.csv")

# Select only useful columns
fao_data_clean = fao_data[[
    "Area", "Year", "Item", "Element", "Unit", "Value"
]].copy()

# Rename for consistency with other datasets
fao_data_clean = fao_data_clean.rename(columns={
    "Area": "Country Name"
})

# Optional: Clean up countries (if needed) or drop missing
fao_data_clean = fao_data_clean.dropna(subset=["Country Name", "Year", "Value"])

# Preview
print(fao_data_clean.head())
```

```python
import pandas as pd

# Load the Tariff dataset
tariff_data = pd.read_csv("data/tariff_data.csv")

# Rename 'Entity' to 'Country Name' for consistency
tariff_data = tariff_data.rename(columns={
    "Entity": "Country Name",
    "Tariff rate, applied, weighted mean, all products (%)": "Tariff Rate (%)"
})

# Optional: Drop rows with missing data in 'Country Name' or 'Year' (if needed)
tariff_data = tariff_data.dropna(subset=["Country Name", "Year"])

# Preview
print(tariff_data.head())
```

```python
import pandas as pd

# Load the dataset
sixth_data = pd.read_csv("data/sixth_data.csv")

# Melt the data to long format (years as rows, instead of columns)
sixth_data_clean = pd.melt(sixth_data,
                           id_vars=["Area", "Item", "Element", "Unit"],
                           value_vars=[f"Y{year}" for year in range(1991, 2025)],
                           var_name="Year", value_name="Value")

# Remove "Y" from year column
sixth_data_clean["Year"] = sixth_data_clean["Year"].str[1:].astype(int)

# Rename for consistency with other datasets
sixth_data_clean = sixth_data_clean.rename(columns={
    "Area": "Country Name",
    "Item": "Item Name",
    "Element": "Element Name"
})

# Optional: Drop rows with missing 'Country Name', 'Year', or 'Value'
sixth_data_clean = sixth_data_clean.dropna(subset=["Country Name", "Year", "Value"])

# Preview the cleaned data
print(sixth_data_clean.head())
```

```python
# Merge all six datasets based on 'Country Name' and 'Year'
final_df = gdp_long.merge(inflation_long, on=["Country Name", "Year"], how="outer")
final_df = final_df.merge(deflation_long, on=["Country Name", "Year"], how="outer")
final_df = final_df.merge(fao_data_clean, on=["Country Name", "Year"], how="outer")
final_df = final_df.merge(tariff_data, on=["Country Name", "Year"], how="outer")
final_df = final_df.merge(sixth_data_clean, on=["Country Name", "Year"], how="outer")

# Preview the final merged dataset
#print(final_df.head())

```
